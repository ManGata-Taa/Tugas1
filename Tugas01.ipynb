Membuat array untuk menyimpan nilai

```python
data_nilai = []
```

Program untuk memasukkan ke data

```python
for i in range(10):
    angka = int(input(f"Input nilai ke-{i+1}: "))
    data_nilai.append(angka)
```

Menghitung jumlah mahasiswa yang lulus

```python

jumlah_lulus = 0
for x in data_nilai:
    if x >= 60:
        jumlah_lulus += 1
```

Menampilkan hasil

```python
print("===== HASIL PERHITUNGAN =====")
print("Data Nilai        :", data_nilai)
print("Nilai Maksimum    :", max(data_nilai))
print("Nilai Minimum     :", min(data_nilai))
print("Rata-rata Nilai   :", sum(data_nilai)/len(data_nilai))
print("Total Lulus       :", jumlah_lulus)
print("Total Tidak Lulus :", len(data_nilai) - jumlah_lulus)
```
