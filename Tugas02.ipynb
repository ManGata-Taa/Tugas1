Nilai Array

```python
import numpy as np

# Membuat array yang berisi kumpulan nilai
array_nilai = np.array([40, 70, 95, 60, 55, 85, 90, 100, 80, 75])
print("Isi Array Nilai :", array_nilai)

# Mencari nilai tertinggi dan terendah dari array
nilai_tertinggi = max(array_nilai)
nilai_terendah = min(array_nilai)

# Menghitung jumlah yang lulus
jumlah_lulus = 0
for nilai in array_nilai:
    if nilai >= 60:
        jumlah_lulus += 1

# Menghitung jumlah yang tidak lulus
jumlah_tidak_lulus = len(array_nilai) - jumlah_lulus
```

Grafik Perbandingan Nilai

```python
import matplotlib.pyplot as plt

# Label untuk grafik batang
label_grafik = ["Nilai Tertinggi", "Nilai Terendah"]
data_grafik = [nilai_tertinggi, nilai_terendah]

plt.bar(label_grafik, data_grafik)
plt.title("Grafik Perbandingan Nilai")
plt.ylabel("Nilai")
plt.show()
```

Grafik Persentase Kelulusan

```python

import matplotlib.pyplot as plt

# Data kelulusan
label_pie = ["Lulus", "Tidak Lulus"]
data_pie = [jumlah_lulus, jumlah_tidak_lulus]

plt.pie(data_pie, labels=label_pie, autopct='%1.1f%%')
plt.title("Grafik Persentase Kelulusan")
plt.axis('equal')
plt.show()
```
